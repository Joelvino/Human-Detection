```python
def statistic_prop(signals,waveletname,lev,n):
    feature_mean=[]; feature_variance=[];feature_std=[];feature_max=[];feature_min=[];feature_median=[];feature_quartile=[];feature_third_quartile=[]; feature_energ=[];feature_rms=[]
    for i in range(40):
        for j in range(200):
            coeff=pywt.wavedec(signals[i][j],waveletname,level=lev)
            feature_mean.append(np.mean(coeff[0]))
            feature_variance.append(np.var(coeff[0]))
            feature_std.append(np.std(coeff[0]))
            feature_min.append(np.min(coeff[0]))
            feature_max.append(np.max(coeff[0]))
            feature_rms.append(np.mean(np.sqrt(coeff[0]**2 + coeff[1]**2)))
            feature_median.append(np.percentile(coeff[0], 50))
            feature_quartile.append(np.percentile(coeff[0], 25))
            feature_third_quartile.append(np.percentile(coeff[0], 75))
            feature_energ.append(sum(coeff[0]**2 + coeff[1]**2))
    labels=[]
    for j in range(8000):
        labels.append(n)
    return pd.DataFrame(np.column_stack([feature_mean,feature_variance,feature_rms, feature_std, feature_min, feature_max,feature_median,feature_quartile, feature_third_quartile, feature_energ, labels]), 
                               columns=['Mean','Variance','RMS','Std', 'MIN','MAX','Median', 'Quartile','Third_quartile', 'Energy','number of people'])
```
